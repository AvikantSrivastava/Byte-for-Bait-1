```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import cv2
import numpy as np
```

```python
pic = cv2.imread('ff_6.jpg')  # dividing by 255 to bring the pixel values between 0 and 1
print(pic.shape)
plt.figure(figsize=(7,7))
plt.imshow(pic)
```

```python
img=cv2.cvtColor(pic,cv2.COLOR_BGR2RGB)
```

```python
pic_n = img.reshape(img.shape[0]*img.shape[1], img.shape[2])
pic_n = np.float32(pic_n)
```

```python
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
K = 4
attempts=10
ret,label,center=cv2.kmeans(pic_n,K,None,criteria,attempts,cv2.KMEANS_PP_CENTERS)
```

```python
center = np.uint8(center)
res = center[label.flatten()]
result_image = res.reshape((img.shape))
# plt.imshow(cluster_pic)
```

```python
f, pos = plt.subplots(1, 2, figsize=(30,30))
# plt.figure(figsize=(10,10))
pos[0].imshow(pic)
pos[0].set_title("Before")
pos[0].axis('off')

pos[1].imshow(result_image)
pos[1].set_title("After")
pos[1].axis('off')

plt.show()
```
